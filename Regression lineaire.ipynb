#  Régression des moindres carrés ordinaires

```python
Boston_numpy=datasets.load_boston(True)
Boston,y_Boston=Boston_numpy[0],Boston_numpy[1]
Diabetes_numpy=datasets.load_diabetes(True)
Diabetes,y_Diabetes=Diabetes_numpy[0],Diabetes_numpy[1]
def App_Matrix(X):
    X=np.append(X, np.ones([X.shape[0], 1], dtype=np.int32), axis=1)
    return X
X_Boston=App_Matrix(Boston)
X_Diabetes=App_Matrix(Diabetes)
```

Cette fonction Crée la matrice d’apprentissage X telle que nous l’avons vu en cours, en ajoutant une colonne
de 1 aux exemples

```python
def Get_W_b(X,y):
    alpha = np.dot(np.linalg.inv(np.dot(np.transpose(X),X)),np.dot(np.transpose(X),y))
    b=alpha[alpha.size-1]
    w=alpha[:-1].copy()
    prediction=X@alpha
    return prediction,w,b
```

cette fonction crée la variable alpha en utilisant l'équation que nous avons vue en classe, puis le biais est la dernière valeur dans le tableau alpha et les poids de la fonction sont les autres valeurs.ensuite nous avons fait les prédictions en multipliant alpha avec l'ensemble de données X.

```python
Boston_pred,w_boston,b_boston=Get_W_b(X_Boston,y_Boston)
```

```python
def plotGraph(y_test,y_pred,regressorName):
    if max(y_test) >= max(y_pred):
        my_range = int(max(y_test))
    else:
        my_range = int(max(y_pred))
    plt.scatter(range(len(y_test)), y_test, color='blue')
    plt.scatter(range(len(y_pred)), y_pred, color='red')
    plt.title(regressorName)
    plt.show()
    return

plotGraph(y_Boston, Boston_pred, "difference between actual and predicted values")
```

nous pouvons voir qu'il y a une petite différence entre nos valeurs prédites et les valeurs réelles et cette conclusion sera plus fiable lorsque nous verrons le mse et le coefficient r2

```python
from sklearn.metrics import mean_squared_error,r2_score
def calculate_metrics(y_true,y_pred):
    print("the mean squared error is:",mean_squared_error(y_true, y_pred))
    print("the r2 is:",r2_score(y_true, y_pred))
calculate_metrics(y_Boston,Boston_pred)
```

we can see from these values that there is a difference between what we are predicting and the real values. we will see how to solve this problem after.

maintenant pour l'ensemble de données sur le diabetes

```python
Diab_pred,w_diab,b_diab=Get_W_b(X_Diabetes,y_Diabetes)
plotGraph(y_Diabetes,Diab_pred,"difference between actual and predicted values")
```

```python
calculate_metrics(y_Diabetes,Diab_pred)
```

# Régression Ridge

```python
Boston_train, Boston_test, y_Boston_train, y_Boston_test = train_test_split(X_Boston, y_Boston, test_size=0.3, random_state=42)
Diabetes_train, Diabetes_test, y_Diabetes_train, y_Diabetes_test = train_test_split(X_Diabetes, y_Diabetes, test_size=0.3, random_state=42)
```

```python
def regression_ridge(X,y,lambda_R):
    S=np.eye(X.shape[1],dtype=int)
    S[X.shape[1]-1,X.shape[1]-1]=0
    alpha = np.dot(np.linalg.inv(np.dot(np.transpose(X),X)+S*lambda_R),np.dot(np.transpose(X),y))
    return alpha
def predict(X,alpha):
    return X@alpha
```

Cette fonction prend une valeur Lambda et nous renvoie l'alpha correspondant à ce lambda, et la fonction Predict renvoie les Y prédits pour cet alpha.

Nous allons maintenant essayer ce modèle sur le jeu de données de Boston.


```python
Lambda=np.logspace(-4,2,20)
Mse_Train=[]
Mse_Test=[]
R2_Train=[]
R2_test=[]
for i in range(len(Lambda)):
    alpha=regression_ridge(Boston_train,y_Boston_train,Lambda[i])
    train_pred=predict(Boston_train,alpha)
    test_pred=predict(Boston_test,alpha)
    Mse_Train.append(mean_squared_error(y_Boston_train,train_pred))
    Mse_Test.append(mean_squared_error(y_Boston_test,test_pred))
    R2_Train.append(r2_score(y_Boston_train,train_pred))
    R2_test.append(r2_score(y_Boston_test,test_pred))
    
```

Nous avons maintenant entre les mains l'erreur quadratique moyenne sur les ensembles de données de formation et de test (Boston) pour chaque valeur de lambda et nous allons les représenter graphiquement.

```python
# Plot the mean train error and test error 

plt.figure(figsize=(6, 4))

plt.semilogx(Lambda,Mse_Train,label='Mse Train')
plt.semilogx(Lambda, Mse_Test,
         label='Mse Test')
plt.legend()
plt.xlabel('Lambda')
plt.ylabel('Mse')

```

```python
# Plot the r2 score on train and test 
plt.figure(figsize=(6, 4))
plt.semilogx(Lambda, R2_test,
         label='R2 Test')
plt.semilogx(Lambda,R2_Train,label='R2 Train')

plt.legend(loc='best')
plt.xlabel('Lambda')
plt.ylabel('R2')
#plt.tight_layout()
```

nous pouvons voir que lorsque nous augmentons lambda, l'erreur augmente et le R2 diminue, nous devons donc utiliser une petite valeur de lambda.

Nous allons maintenant voir les résultats sur l'ensemble de données sur le diabetes.

```python
Lambda=np.logspace(-4,2,20)
Mse_Train=[]
Mse_Test=[]
R2_Train=[]
R2_test=[]
for i in range(len(Lambda)):
    alpha=regression_ridge(Diabetes_train,y_Diabetes_train,Lambda[i])
    train_pred=predict(Diabetes_train,alpha)
    test_pred=predict(Diabetes_test,alpha)
    Mse_Train.append(mean_squared_error(y_Diabetes_train,train_pred))
    Mse_Test.append(mean_squared_error(y_Diabetes_test,test_pred))
    R2_Train.append(r2_score(y_Diabetes_train,train_pred))
    R2_test.append(r2_score(y_Diabetes_test,test_pred))
    
```

```python
plt.figure(figsize=(6, 4))

plt.semilogx(Lambda,Mse_Train,label='Mse Train')
plt.semilogx(Lambda, Mse_Test,
         label='Mse Test')
plt.legend()
plt.xlabel('Lambda')
plt.ylabel('Mse')

```

```python
# Plot the r2 score on train and test 
plt.figure(figsize=(6, 4))
plt.semilogx(Lambda, R2_test,
         label='R2 Test')
plt.semilogx(Lambda,R2_Train,label='R2 Train')

plt.legend(loc='best')
plt.xlabel('Lambda')
plt.ylabel('R2')
#plt.tight_layout()
```

nous pouvons voir que lorsque nous augmentons lambda, l'erreur augmente et le R2 diminue, donc nous devrions utiliser une petite valeur de lambda. donc même résultat que le jeu de données de Boston. 

On constate également que les valeurs de la mse et du R2 pour le plus petit lambda sont meilleures que celles de la régression linéaire normale

# Normalization et reduction

```python
Boston_train, Boston_test, y_Boston_train, y_Boston_test = train_test_split(Boston, y_Boston, test_size=0.3, random_state=42)
Diabetes_train, Diabetes_test, y_Diabetes_train, y_Diabetes_test = train_test_split(Diabetes, y_Diabetes, test_size=0.3, random_state=42)
```

```python
train_mean=Boston_train.mean(axis=0)
std_Boston_train = np.std(Boston_train, axis = 0)
Boston_train_centered=Boston_train-train_mean
Boston_train_centered_red=Boston_train_centered/std_Boston_train
Boston_train_centered=App_Matrix(Boston_train_centered)
ytrain_mean=y_Boston_train.mean(axis=0)
y_boston_train_centered=y_Boston_train-ytrain_mean
y_Boston_test_centered=y_Boston_test-ytrain_mean
Boston_test_centered=Boston_test-train_mean
Boston_test_centered_red=Boston_test_centered/std_Boston_train
Boston_test_centered=App_Matrix(Boston_test_centered)
Y_std=np.std(y_Boston_train,axis=0)
y_train_cent_red=y_boston_train_centered/Y_std
y_test_centred_red=y_Boston_test_centered/Y_std
```

```python
lamda=0.00000001
alpha_Centered=regression_ridge(Boston_train_centered,y_boston_train_centered,lamda)
print("le bias de modele centree avec reponse centree est: ", alpha_Centered[len(alpha_Centered)-1])
```

```python
alpha_non_centered=regression_ridge(App_Matrix(Boston_train),y_Boston_train,lamda)
print("le bias de modele non centree est: ", alpha_non_centered[len(alpha_non_centered)-1])
```

```python
lamda=0.00000001
alpha_Centered_ynot=regression_ridge(Boston_train_centered,y_Boston_train,lamda)
print("le bias de modele centree avec reponse noncentree est: ", alpha_Centered_ynot[len(alpha_Centered_ynot)-1])
```

```python
lamda=0.00000001
alpha_Centered_red=regression_ridge(Boston_train_centered_red,y_train_cent_red,lamda)
print("le bias de modele centree et reduit: ", alpha_Centered_red[len(alpha_Centered_red)-1])
```

```python
print("moyenne des réponses de l’ensemble d’apprentissage est: ", y_Boston_train.mean())
```

```python
print("those are metrics for centered dataset")
calculate_metrics(y_Boston_test_centered,predict(Boston_test_centered,alpha_Centered))
```

```python
print("those are the metrics for non centered dataset")
calculate_metrics(y_Boston_test,predict(App_Matrix(Boston_test),alpha_non_centered))
```

```python
print("those are the metrics for centered and reduit dataset and centered and reduite y")
calculate_metrics(y_test_centred_red,predict(Boston_test_centered_red,alpha_Centered_red))
```

nous avons remarqué qu'il y a une petite amélioration lorsque l'ensemble de données est centré et réduit (r2 augmenté), cependant lorsqu'il est centré seulement il n'y a pas d'amélioration.
